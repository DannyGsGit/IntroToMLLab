```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(color_codes=True)
import pylab
import random
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.preprocessing import scale
import xgboost as xgb
import os
```

```python
def clusterGroups(df):
    # Get array from major summaries
    X = scale(np.array(df))
    levels = df.index
    # Prep result lists
    clust_list = []
    result_list = []
    labels_list = []
    # Run multiple kmeans with different counts of clusters
    for n_clust in range(2, len(df)):
        kmeans_model = KMeans(n_clusters=n_clust, random_state=1).fit(X)
        labels = kmeans_model.labels_
        labels_list.append(labels)
        clust_list.append(n_clust)
        # Calinski Harabaz metric will be used to choose "best" number of clusters
        result_list.append(metrics.calinski_harabaz_score(X, labels))

    # Get labels for "best" clustering result
    clusters = labels_list[np.argmax(result_list)]

    # Create mapping table
    cluster_maps = pd.DataFrame({'original': levels, 'clusters': clusters})
    ## Rename clusters
    cluster_maps['clusters'] = 'CL' + cluster_maps['clusters'].astype('str')
    return(cluster_maps)



# UDFs for quantile computations
def q25(x):
    return x.quantile(0.25)
def q75(x):
    return x.quantile(0.75)

# Function to get mean/IQR tables
def getMeanIQRTable(df, target, aggfeature):
    # Calculate summary stats for each job type
    summary_table = df[[target, aggfeature]].groupby(aggfeature).agg({target: ['mean', q25, q75]}).reset_index()
    # Calculate IQRs and Upper/Lower bounds
    summary_table['IQR']= summary_table[target]['q75'] - summary_table[target]['q25']
    summary_table['LowerBound']= summary_table[target]['q25'] - (1.5 * summary_table['IQR'])
    summary_table['UpperBound']= summary_table[target]['q75'] + (1.5 * summary_table['IQR'])
    # Relabel columns to flatten
    summary_table.columns = [''.join(col).strip() for col in summary_table.columns.values]
    # Downselect columns and re-order
    summary_table = summary_table[[aggfeature, '{}mean'.format(target), 'LowerBound', 'UpperBound']]
    summary_table.sort_values('{}mean'.format(target), inplace=True)
    return(summary_table)



def data_prep(df, jobType_table, major_table):
    # Re-map jobtypes
    df.replace({'jobType': {'CTO': 'CXO', 'CFO': 'CXO'}}, inplace=True)
    
    # Join jobtype/major tables
    df = df.merge(jobType_table, on='jobType', how='left')
    df = df.merge(major_table, on='major', how='left')
    
    # Cleanup
    df.drop(['LowerBound', 'UpperBound'], axis=1, inplace=True)
    return(df)



def rank_associations(df, targetCol, catList, subsample=0.2):
    # Speed things up by subsampling
    df = df.sample(frac=subsample, replace=False)
    
    # Prepare the plot
    plt.figure(figsize=(10,10))
    
    # Interaction with categorical variables
    etasquared_dict = {}
    if len(catList) >= 1:
        for each in catList:
            mod = ols('{} ~ C({})'.format(targetCol, each), data=df[[targetCol,each]]).fit()
            aov_table = sm.stats.anova_lm(mod, typ=1)
            esq_sm = aov_table['sum_sq'][0]/(aov_table['sum_sq'][0]+aov_table['sum_sq'][1])
            etasquared_dict[each] = esq_sm

        topk_esq = pd.DataFrame.\
                    from_dict(etasquared_dict, orient='index').\
                    unstack().\
                    sort_values(kind = 'quicksort', ascending=False).\
                    head(25).\
                    reset_index().\
                    set_index('level_1')
        topk_esq.columns = ['level_0', 'EtaSquared']
        topk_esq[['EtaSquared']].plot(kind='bar')
        plt.xlabel('Feature')
        plt.ylabel('Eta^2')
        plt.title('Eta Square Values')
    return(topk_esq)




def residual_plots(df, sample_fraction):
    # Subsample plot data
    df = df.sample(frac=sample_fraction, replace=False)
    
    # Residual plot
    f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 12))
    ax1.scatter('pred', 'residual', data=df, alpha=0.1, s=1)
    ax1.axhline(y=0, c='k')
    ax1.set_xlabel('Predicted')
    ax1.set_ylabel('Residual')

    # Pred-actual plot
    ax2.scatter('y', 'pred', data=df, alpha=0.1, s=1)
    ax2.plot((0,300), (0,300), ls="--", c=".3")
    ax2.set_xlabel('Actual')
    ax2.set_ylabel('Predicted')
    
    
    # Histogram of residuals
    ax3.hist(df['residual'], bins = 20)
    ax3.set_xlabel('Residual')
    
    # Distributions of salaries
    sns.kdeplot(df['pred'], ax=ax4)
    sns.kdeplot(df['y'], ax=ax4)
    
    plt.show()

    
    
    
def one_hot_mixed_df(data):
    data_cat = data.select_dtypes(include='object')
    data_num = data.select_dtypes(exclude='object')
    ## One-hot transform the objects
    data_cat_dummies = pd.get_dummies(data_cat,drop_first=True)
    ## Join dummies with numerics
    data_lm = pd.concat([data_cat_dummies, data_num], axis=1, sort=False)
    return(data_lm)
```

### Target Transformation

By visual inspection, salary looks a bit skewed. Given our sample size a Shapiro test won't work, but inspection of qq-plots can inform any transformation of the salary.

```python
# Generate multiple transformations using this snippet from the SK-learn examples 
# (http://scikit-learn.org/stable/auto_examples/preprocessing/plot_all_scaling.html#sphx-glr-auto-examples-preprocessing-plot-all-scaling-py)

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import QuantileTransformer
from sklearn.preprocessing import PowerTransformer

y = np.array(train_data['salary'].astype('float64'))
y = y.reshape(-1,1)

distributions = [
    ('unscaled data', y),
    ('data after standard scaling',
        StandardScaler().fit_transform(y)),
    ('data after robust scaling',
        RobustScaler(quantile_range=(25, 75)).fit_transform(y)),
    ('data after power transformation (Yeo-Johnson)',
        PowerTransformer(method='yeo-johnson').fit_transform(y)),
    ('data after power transformation (Box-Cox)',
        PowerTransformer(method='box-cox').fit_transform(y)),
    ('data after quantile transformation (gaussian pdf)',
        QuantileTransformer(output_distribution='normal').fit_transform(y))
]
```

```python
for i in range(0, len(distributions)):
    # Reshape distribution into a pandas series for plotting/testing
    y = pd.Series(distributions[(i)][1].reshape(1, -1)[0])
    # Generate QQ and distribution plots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
    fig.suptitle('QQ and Distribution Plots for {}'.format(distributions[(i)][0]))
    sm.qqplot(y, line='s', ax=ax1)
    sns.distplot(y, ax=ax2)
    pylab.show()  # Needed for the QQ plot
```

Examining the outputs of the scalers, we see that power transformations in particular help shift the salary distribution toward a normal distribution with the <span style="color:red">**Yeo-Johnson Power Transformation**</span> resulting in a slightly smoother distribution. We will use this transformation for our target variable in training.

```python
pt = PowerTransformer(method='yeo-johnson', standardize=False)
# Shape salary for the transformer
y = np.array(train_data['salary'].astype('float64'))
y = y.reshape(-1,1)

# Fit transformation
pt.fit(y)

# Apply transformation to salary
train_data['salary_xform'] = pt.transform(y)[:, 0]
# Reverse the transformation later with: pt.inverse_transform(np.array(train_data['salary_xform']).reshape(-1, 1))
```

## Model

```python
# Split our data columns into features and target tables
cols_to_drop = ['salary', 'salary_xform', 'salary_log', 'jobId', 'companyId', 'milesFromMetropolis']
target = 'salary'
X_train = train_data.drop(columns=cols_to_drop)
y_train = train_data[[(target)]].astype('float')

# Apply scaling
scaler = StandardScaler()
scaler.fit(X_train.select_dtypes(include='number'))
X_train[list(X_train.select_dtypes(include='number'))] = scaler.fit_transform(X_train.select_dtypes(include='number'))

# Prep test split
X_test = eval_data.drop(columns=cols_to_drop)
y_test = eval_data[[(target)]].astype('float')
X_test[list(X_train.select_dtypes(include='number'))] = scaler.fit_transform(X_test.select_dtypes(include='number'))

# One-hot encode any categorical variables
# Apply transformations
X_train_onehot = one_hot_mixed_df(X_train)
X_test_onehot = one_hot_mixed_df(X_test)
```

#### XG Boost
We will begin by running a gridsearch for parameter optimization. This phase will be run on a subsample of the data, with an aggressive learning rate and fewer estimators to speed up the process. When optimized parameters have been selected, we will back off learning rate and train with more estimators.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold

# Generate a subsample for quicker training
sub_idx = np.random.randint(0, len(y_train), int(len(y_train)*0.3))
X_train_cv = X_train_onehot.iloc[sub_idx,:]
y_train_cv = y_train.iloc[sub_idx,:]

xg_reg = xgb.XGBRegressor(objective ='reg:linear',
                          learning_rate = 0.3,
                          n_estimators = 25)

colsample_bytree = [0.3, 0.6, 0.8]
max_depth = [2, 4, 6, 8]
alpha = [1, 3, 8]
subsample = [0.3, 0.6, 0.8]
param_grid = dict(colsample_bytree=colsample_bytree,
                  max_depth=max_depth,
                  reg_alpha=alpha,
                  subsample=subsample)

kfold = KFold(n_splits=10, shuffle=True, random_state=7)
grid_search = GridSearchCV(xg_reg, param_grid, scoring="neg_mean_squared_error", n_jobs=-1, cv=kfold, verbose=1)
grid_result = grid_search.fit(X_train_cv, y_train_cv)
```

```python
print("Best Score (MSE): {} \nParams: {}".format(grid_result.best_score_, grid_result.best_params_))
```

Now it's time to train more conservatively. Using the parameters determined from the CV, we will re-train the model with a lower learning rate and more trees for a "slow burn" training session.

```python
xg_reg = xgb.XGBRegressor(objective ='reg:linear',
                          colsample_bytree = 0.6,
                          learning_rate = 0.1,
                          max_depth = 8,
                          subsample=0.8,
                          alpha = 3,
                          n_estimators = 500)

# Generate weights to adjust heteroscedasticity
mx = train_data[target].max()
weights = mx/train_data[target]

# Train model
xgbData = xgb.DMatrix(X_train_onehot, label=y_train)
eval_set = [(X_train_onehot, y_train), (X_test_onehot, y_test)]
xg_reg.fit(X_train_onehot, y_train, sample_weight=weights, eval_metric="rmse", eval_set=eval_set, verbose=False)
```
